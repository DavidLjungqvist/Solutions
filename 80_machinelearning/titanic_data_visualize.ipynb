```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("titanic_train.csv", index_col="PassengerId")
# Git saved my life
df
```

```python
df["CabinDeck"] = df["Cabin"].str[0]
df
```

```python
df["Ticket"].value_counts()
```

```python
df["Title"] = df["Name"].str.extract(r',\s*([^\.]+)\.')
title_survival = df.groupby("Title")["Survived"].mean().sort_values()
title_survival
```

```python
df["Title"].value_counts()
```

```python
df["Title"] = df["Title"].replace({"Mlle": "Miss", "Mme": "Mrs", "Ms": "Miss"})
high_status = ["the Countess", "Sir", "Lady", "Don", "Dona", "Jonkheer"]
above_average_title = ["Dr", "Major", "Col"]
hardcore_mfs = ["Capt", "Rev"]
df["Title"] = df["Title"].replace(high_status, "High Status")
df["Title"] = df["Title"].replace(above_average_title, "Improved Status")
df["Title"] = df["Title"].replace(hardcore_mfs, "Mr")
title_survival = df.groupby("Title")["Survived"].mean().sort_values()
title_survival
```

```python
test = pd.read_csv("titanic_test.csv")
test["Title"] = test["Name"].str.extract(r',\s*([^\.]+)\.')
test["Title"] = test["Title"].replace({"Mlle": "Miss", "Mme": "Mrs", "Ms": "Miss"})
high_status = ["the Countess", "Sir", "Lady", "Don", "Dona", "Jonkheer"]
above_average_title = ["Dr", "Major", "Col"]
hardcore_mfs = ["Capt", "Rev"]
test["Title"] = test["Title"].replace(high_status, "High Status")
test["Title"] = test["Title"].replace(above_average_title, "Improved Status")
test["Title"] = test["Title"].replace(hardcore_mfs, "Mr")
print(test["Title"].unique())
print(test["Title"].value_counts())
```

```python
# title_map = {title: i for i, title in enumerate(title_survival.index)}
title_map = {'Mr': 0,
 'Improved Status': 1,
 'Master': 3,
 'High Status': 2,
 'Miss': 4,
 'Mrs': 5}
title_map
```

```python
gender_map = {"male": 0, "female": 1}
df["Gender_encoded"] = df["Sex"].map(gender_map)
```

```python
sns.stripplot(data=df, x="Pclass", y="Fare", jitter=True)
plt.title("Pclass vs Fare (Jittered)")
plt.show()
```

```python
sns.violinplot(data=df, x="Pclass", y="Fare")
plt.title("Fare Distribution by Pclass (Violin)")
plt.show()
```

```python
pivot = df.pivot_table(
    values="Survived",
    index="SibSp",
    columns="Parch",
    aggfunc="mean"
)

sns.heatmap(pivot, annot=True, cmap="coolwarm")
plt.title("Mean Survival Rate by SibSp & Parch")
plt.show()
```
